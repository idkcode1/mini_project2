```python
import tensorflow as tf
import random
from nltk.tokenize import word_tokenize
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.optimizers import Adam
from datetime import datetime
```

```python
intents = {
    "greeting": ["hello", "hi", "hey", "hiya"],
    "farewell": ["bye", "goodbye", "see you", "take care",'thank you', 'thanks', 'thanks a lot', 'appreciate it'],
    "weather": ["what is the weather", "how is the weather", "weather today"],
    "time": ["what is the time", "current time", "time now"],
    "commands": ['open the file', 'close the file', 'save the file', 'copy', 'paste', 'undo', 'redo'],
    "age": ['how old are you?', 'what is your age?'],
    "google_search": ['search for', 'google', 'search','chad bot'],
}

responses = {
    "greeting": ['Hello!', 'Hi there!', 'Hey!', 'How can I help you?'],
    "farewell": ['thank you','Goodbye!', 'Take care!', 'See you later!'],
    "commands": ['The requested task is being performed'],
    "age": ['I don\'t have an age. I\'m just a computer program!', 'I don\'t really age, I just get updated with new features and improvements.'], 
    "weather": ['The weather is currently', 'I am not sure, would you like me to check?'],
    "time": ['The time is currently', 'It is currently'],
    "thanks": ['You\'re welcome!', 'No problem!', 'Anytime!'],
    "google_search": ['Searching Google', 'Let me do it for you'],
    "unknown": ['I\'m sorry, I don\'t understand.', 'Could you please rephrase your question?', 'I\'m not sure what you\'re asking. Can you provide more context?', 'I\'m not familiar with that command. Could you explain it to me?']
}
```

```python
words = []
classes = []
documents = []
for intent, intent_words in intents.items():
    for word in intent_words:
        words.extend(word_tokenize(word))
        documents.append((word_tokenize(word), intent))
        if intent not in classes:
            classes.append(intent)
words = sorted(list(set(words)))
```

```python
import numpy as np
X = []
y = []
for document in documents:
    intent = document[1]
    bag = []
    for word in words:
        bag.append(1) if word in document[0] else bag.append(0)
    X.append(bag)
    y.append(classes.index(intent))
X = tf.keras.utils.normalize(np.array(X), axis=1)
y = np.array(y)
```

```python
# Build the model
model = Sequential()
model.add(Dense(128, input_shape=(len(words),), activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(64, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(len(classes), activation='softmax'))
optimizer = Adam(lr=0.001)
model.compile(optimizer=optimizer, loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
model.fit(X, y, epochs=200, batch_size=5)
```

```python
def predict_intent(sentence):
    bag = []
    for word in words:
        bag.append(1) if word in word_tokenize(sentence) else bag.append(0)
    bag = np.array(bag)
    result = model.predict(np.array([bag]),verbose=0)[0]
    intent_idx = np.argmax(result)
    intent = classes[intent_idx]
    return intent
```
